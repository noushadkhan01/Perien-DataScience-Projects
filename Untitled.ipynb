```python
import numpy as np, seaborn as sns, pandas as pd, matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
tips.info()
```

```python
g = sns.lmplot(x = 'tip', y = 'total_bill', data = tips, aspect = 2, hue = 'sex'
              )
g = g.set_axis_labels('tip', 'Total Bill(US $)').set(xlim = (0, 12), ylim = (0, 70))
plt.title('Regression plot bw tip and total_bill')
plt.show(g)
```
