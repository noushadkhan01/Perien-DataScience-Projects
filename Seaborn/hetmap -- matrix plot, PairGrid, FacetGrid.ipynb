```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
flights = sns.load_dataset('flights')
flights.head(2)
```

### Note:-- matrix plot works on matrix like data

### example:--

#### to form a matrix data we need to convert data into pivot_table

```python
fp = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
fp
```

## create seaborn heatmap

```python
sns.heatmap(fp, linewidth = 3, robust = True)
plt.xticks(rotation = 50)
```

# sns.PairGrid()

```python
iris = sns.load_dataset('iris')
iris.head(2)
```

```python
sns.pairplot(iris, hue = 'species', #kind = 'reg'
            )
```

```python
sns.PairGrid(iris)
```

#### ^As you can see it's an empty plot but it's very useful for further customize it

### PairGrid.map(kind = 'plotType')

```python
g = sns.PairGrid(iris, hue = 'species')
g.map(plt.scatter) #this will change diagonal plots
```

```python
sns.distplot()
```

```python
g = sns.PairGrid(iris, hue = 'species')
g.map_diag(sns.distplot, rug = True)
g.map_upper(plt.scatter)
g.map_lower(sns.jointplot, kind = 'reg')
```

## FacetGrid()

```python
tips.head(2)
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker' )
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker' )
g.map(sns.distplot, 'total_bill', rug = True)
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker' )
g.map(sns.jointplot, 'total_bill', 'tip', kind = 'reg')
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker' )
g.map(plt.scatter, 'total_bill', 'tip')
```
